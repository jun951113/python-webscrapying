```python
import pandas as pd
import numpy

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dy.csv')
```

```python
df = df.iloc[:27]
```

```python
df
```

```python
df['name'] = df['name'].str.replace('INTO1–刘宇：', ' ')
```

```python
df['like'] = df['like'].str.replace('w', ' ').astype(float)
```

```python
df.head(10)
```

```python
plt.plot(df['like'])
```

```python
ax = plt.gca()
df['like'].plot(kind='line',x='name',y='likes', color='blue', ax=ax)
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False


x1 = df['name']
 
# 体重
y1 = df['like']
# 设置画布大小
fig = plt.figure(figsize=(160, 40))
fig.autofmt_xdate()
 
# 标题
plt.title("dy_data")
 
# 数据
# plt.plot(x1, y1, label='likes changes', linewidth=3, color='r', marker='o',
#          markerfacecolor='blue', markersize=200)
plt.bar(x1, y1, 0.5, color='blue',edgecolor='grey') 
# 横坐标描
plt.xlabel('name',rotation = 90)
# 纵坐标描述
plt.ylabel('likes')
 
# 设置数字标签
for a, b in zip(x1, y1):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=20)
 
plt.savefig('books_read.png')


plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False


fig, ax = plt.subplots(figsize=(160,160))
fig.autofmt_xdate()

y_data = df['like']
x_data = df['name']

# 柱状图颜色
color = 'coral'

# 柱状图

bar = plt.plot(x_data,y_data, label='likes changes', linewidth=3, color='r', marker='o',
         markerfacecolor='blue', markersize=20)

# bar = plt.bar(x_data, y_data, 0.5, color='blue',edgecolor='grey')

# 设置标题
ax.set_title('抖音点赞',fontsize=20,y=1.05)
# 设置坐标轴标题
ax.set_xlabel('name',fontsize = 40)
ax.set_ylabel("点赞数",fontsize = 40,color = 'black',alpha = 0.7,rotation=360)
# 设置Y轴区间
ax.set_ylim(0,30)

# 显示数据标签

for a, b in zip(x_data, y_data):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=40)

plt.savefig('books_read.png')

# 边框隐藏
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
```
